```python
def fetch_videos(query, max_total_results=200):
    all_video_details = []
    request = youtube.search().list(
        q=query,
        part='id,snippet',
        maxResults=50,  # Maximum allowed by the API in a single request
        type='video'
    )

    while request is not None and len(all_video_details) < max_total_results:
        response = request.execute()
        for item in response['items']:
            video_details = {
                'videoId': item['id']['videoId'],
                'title': item['snippet']['title'],
                'description': item['snippet']['description'],
                'channelTitle': item['snippet']['channelTitle']
            }
            all_video_details.append(video_details)
            if len(all_video_details) >= max_total_results:
                break

        # Check if there is a nextPageToken and if the maximum desired total results have not been reached
        if 'nextPageToken' in response and len(all_video_details) < max_total_results:
            request = youtube.search().list_next(request, response)
        else:
            request = None

    return all_video_details

# Fetch videos
video_details = fetch_videos('tech review', max_total_results=200)
```
